```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['Preemptive-Priority','Round Robin','Optimized Priority']
revenue=[9.67,9.33,5.33]
profit=[6.5,12.67,8.67]
art=[7.33,3.33,6.33]
xpos=np.arange(0,len(company)*2,2)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos, revenue, label="AWT",width = 0.4)
plt.ylabel("Time")
plt.xlabel("Algorithm")
plt.title('Comparison')
plt.bar(xpos+0.4,profit, label = 'ATT',width = 0.4)
plt.bar(xpos+0.8,art, label = 'ART',width = 0.4)
plt.legend()
plt.savefig("piechart.png", bbox_inches='tight', pad_inches=1, transparent=True)
```
