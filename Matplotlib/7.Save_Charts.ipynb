```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%')
plt.savefig("piechart.png", bbox_inches='tight', pad_inches=1, transparent=True)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%')
plt.savefig("piechart.pdf", bbox_inches='tight', pad_inches=1, transparent=True)
```
