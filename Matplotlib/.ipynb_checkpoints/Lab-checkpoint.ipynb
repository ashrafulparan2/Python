```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['FCFS','Non-Preemptive-SJF','Preemptive-SJF']
revenue=[7.67,6,2.5]
profit=[11,6,4.67]
art=[7.67,9.33,8]
xpos=np.arange(0,len(company)*2,2)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos, revenue, label="AWT",width = 0.4)
plt.ylabel("Time")
plt.xlabel("Algorithm")
plt.title('Comparison')
plt.bar(xpos+0.4,profit, label = 'ATT',width = 0.4)
plt.bar(xpos+0.8,art, label = 'ART',width = 0.4)
plt.legend()
```
