```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company, revenue, label="Revenue")
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(company, revenue, label="Revenue")
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.bar(company,profit, label = 'Profit')
plt.legend()
```

```python
xpos=np.arange(len(company))
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.bar(xpos-0.2, revenue, label="Revenue", width = 0.4)
plt.bar(xpos+0.2,profit, label = 'Profit', width = 0.4)
plt.legend()
```

# horizontal bar

```python
plt.barh(company, revenue, label="Revenue")
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.legend()
```
