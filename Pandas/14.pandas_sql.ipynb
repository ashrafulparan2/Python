```python
import pandas as pd
import sqlalchemy
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/applications')
```

```python
df = pd.read_sql_table('customer',engine)
df
```

```python
df = pd.read_sql_table('customer',engine, columns=['name', 'phone_number'])
df
```

```python
df = pd.read_sql_table('orders',engine)
df
```

```python
query = '''
 SELECT customer.name, customer.phone_number, orders.name, orders.amount
 FROM customer INNER JOIN orders
 ON customer.id=orders.customer_id
'''
df = pd.read_sql_query(query,engine)
df
```
