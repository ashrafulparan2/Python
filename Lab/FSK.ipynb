```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
b = input('Input Bit Stream: ')
n=len(b)
t=np.arange(0,n,0.01)
arr=np.zeros(n)
bw=np.zeros(len(t))

for i in range(n):
    if b[i]== '0':
        arr[i]=-1
    else:
        arr[i]=1
    bw[i*100:(i+1)*100]=arr[i]
wo = 2*((2*np.pi*t))
W = wo/2
sinHt = np.sin(wo+W)
sinLt = np.sin(wo-W)
st=np.sin(wo+bw*W)
plt.subplot(7,1,1)
plt.plot(t,bw, color="y")
plt.grid(True)
plt.axis([0,n,-2,2])

plt.subplot(7,1,3)
plt.plot(t,sinHt, color="y")
plt.grid(True)
plt.axis([0,n,-2,2])

plt.subplot(7,1,5)
plt.plot(t,sinLt, color="y")
plt.grid(True)
plt.axis([0,n,-2,2])

plt.subplot(7,1,7)
plt.plot(t,st, color="y")
plt.grid(True)
plt.axis([0,n,-2,2])

Fs=1
plt.figure()
plt.psd(st, color="y")
plt.show()
plt.savefig("output.png", bbox_inches='tight', pad_inches=1, transparent=True)
```
