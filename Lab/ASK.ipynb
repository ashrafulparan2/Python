```python
import numpy as np
import matplotlib.pyplot as plt

b = input('Enter the bit stream: \n')
b = list(map(int, b.split()))

n = len(b)
t = np.arange(0, n+1, 0.01)
x = np.arange(0, (n+2) * 100)

bw = np.zeros(len(x))
i = 0

while i < n+1:
    j = i
    while j < i + 1:
        bw[int(i * 100):int((i + 1) * 100)] = b[i-1]
        j += 0.1
    i += 1

bw = bw[100:]
sint = np.sin(2 * np.pi * t)
st = bw * sint

plt.subplot(3, 1, 1)
plt.plot(t, bw, color='g')
plt.grid(True)
plt.title('Binary Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, sint, color='g')
plt.grid(True)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, st, color='g')
plt.grid(True)
plt.title('Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
