```python
import numpy as np
import matplotlib.pyplot as plt

data = input('Enter the Bit stream: ') 
n = len(data)

t = np.arange(0, n, 0.01)
b_p = np.zeros(n)
bw = np.zeros(len(t))
i = 0
while i < n:
  if data[i] == '0':
    b_p[i] = -1
  else:
    b_p[i] = 1
  bw[i*100: (i+1)*100] = b_p[i]
  i += 1

sint = np.sin(2*np.pi*t)
st = bw * sint

plt.figure(figsize=(8, 6))

plt.subplot(3, 1, 1)
plt.plot(t, bw, color="y")
plt.grid(True)
plt.axis([0, n, -2, 2])
plt.title('Binary Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, sint, color="y")
plt.grid(True)
plt.axis([0, n, -2, 2])
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, st, color="y")
plt.grid(True)
plt.axis([0, n, -2, 2])
plt.title('Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
